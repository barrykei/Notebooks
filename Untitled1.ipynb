```python
import pandas as pd

transactions = {"transaction_id" : [1, 2, 3, 4, 5], "product_id" : [101, 102, 103, 104, 105], "amount" : [3, 5, 8, 3, 2]}

products = {"product_id" : [101, 102, 103, 104, 105], "price" : [20.00, 21.00, 15.00, 16.00, 52.00]}

df_transactions = pd.DataFrame(transactions)

df_products = pd.DataFrame(products)

def transactions_over_100(df_transactions: pd.DataFrame, df_products: pd.DataFrame):
    merged = df_transactions.merge(df_products, on ='product_id')
    merged['total_value'] = merged['amount']*merged['price']
    merged.drop('price', axis = 1, inplace=True)
    merged = merged[merged['total_value'] > 100]
    return merged
```

```python
def is_palindrome_mem(s):
    left = 0
    right = len(s) - 1

    if s[left] == s[right]:
        left += 1
        right -= 1
    else:
        return False
    return True
```

```python
is_palindrome_mem('buchikata')
```

```python
def is_palindrome_cool(s):    
    s_clean = [char.lower() for char in s if char.isalnum()]
    return s_clean == s_clean[::-1]
        
```

```python
is_palindrome_cool('RAcecar')
```

```python
def two_sum(nums, target):
    seen = {}
    
    for i, num in enumerate(nums):
        
        diff = target - num

        if diff in seen:
            return [seen[diff], i]

        seen[num] = i

    return []
            
        
    
    
```

```python
def find_max(nums):
    max_val = nums[0]
    
    for num in nums:
        if num > max_val:
            max_val = num
    return max_val

            

```

```python
find_max([10, 50, 20, 5])
```
