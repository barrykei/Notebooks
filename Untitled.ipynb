```python
def count_animal(amount: int,animal='cow'):
  if amount == 1:
    return 'There is ' + str(amount) + ' ' + animal
  else:
    return 'There are ' + str(amount) + ' ' + animal + 's'

print(count_animal(1))
print(count_animal(2,'dog'))
print(count_animal('one'))
```

```python
class Animal():
  def __init__(self, species, sound):
    self.species = species
    self.sound = sound
  def __str__(self):
    return self.species
  def make_sound(self):
    print(self.sound.capitalize()+'!')

cow = Animal('cow','moo')
print('This is a',cow)
cow.make_sound()
```

```python
def find_bigrams(sentence):
    words = sentence.lower().split()
    return list((word1, word2) for word1, word2 in zip(words, words[1:]))
```
